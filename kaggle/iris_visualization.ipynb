```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
dat=load_iris()
```

```python
print(dat.target_names)
dat.feature_names
```

```python
iris=pd.DataFrame(dat.data, columns=dat.feature_names)
iris['Species']=dat.target
iris.head(2)
```

```python
species_map={0:'setosa',1:'versicolor',2:'virginica'}
iris.Species=iris.Species.map(species_map)
iris.head(3)
```

```python
iris.Species.value_counts()
```

```python
iris.columns
```

```python
iris.plot(kind='scatter',x='sepal length (cm)',y='sepal width (cm)')
```

```python
sns.jointplot('sepal length (cm)','sepal width (cm)',data=iris, size=5)
```

```python
iris.columns=['sepal_length','sepal_width','petal_length','petal_width','species']
```

```python
facet=sns.FacetGrid(iris, hue='species', size=5)
facet.map(plt.scatter, 'sepal_length','sepal_width').add_legend()
```

```python
sns.boxplot(iris.species,iris.petal_length)
```

```python
sns.boxplot(iris.species, iris.petal_length)
sns.stripplot(iris.species,iris.petal_length,jitter=True, edgecolor='gray')
```

```python
sns.violinplot(iris.species,iris.petal_length)
```

```python
sns.FacetGrid(iris, hue='species',size=6)\
.map(sns.kdeplot, 'petal_length')\
.add_legend()
```

```python
sns.pairplot(iris,hue='species',size=3)
```

```python
sns.pairplot(iris, hue='species',size=3, diag_kind='kde')
```

```python
iris.boxplot(by='species',figsize=(8,8))
```

```python
from pandas.tools.plotting import andrews_curves
andrews_curves(iris,'species')
```
